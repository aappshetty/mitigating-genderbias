```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from sacrebleu import corpus_bleu

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-hi-en")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-hi-en")

# Load datasets
file1_path = '/content/drive/MyDrive/total_job_hindi.txt'
file2_path = '/content/drive/MyDrive/total_job_gtrans.txt'

# Read dataset files
with open(file1_path, 'r', encoding='utf-8') as file:
    source_sentences = file.readlines()

with open(file2_path, 'r', encoding='utf-8') as file:
    reference_sentences = file.readlines()

# Tokenize and translate source sentences
tokenized_sentences = tokenizer(source_sentences, truncation=True, padding=True, return_tensors="pt", max_length=512)
translated_sentences = model.generate(**tokenized_sentences)

# Decode translated sentences
translated_texts = tokenizer.batch_decode(translated_sentences, skip_special_tokens=True)

# Calculate BLEU score
bleu_score = corpus_bleu(translated_texts, [reference_sentences])

print("BLEU score:", bleu_score.score)
```
